```python
import random
from itertools import permutations

with open('words7.txt') as f:
    WORDS = sorted(f.read().split(), key=len, reverse =True)

def main():
    clue = list(random.choice([word for word in WORDS if len(word)==7]))
    words = getallwords(clue)
    twist(clue,words)
    
def twist(clue, answers):
    random.shuffle(clue)
    guesses = set()
    printboard(clue, guesses, answers)
    guess = ""
    while len(guesses)<len(answers) and guess != "q":
        guess = input("Guess a word, or q for quit or t for twist: ").lower()
        if guess == "t":
            random.shuffle(clue)
        if guess == "h":
            hint = random.choice(answers)
            #print(hint, "HINT!")
            while hint in guesses:
                hint = random.choice(answers)
                #print("HINT!!1", hint)
            guesses.add(hint)
        else:
            if guess in answers:
                guesses.add(guess)
        printboard(clue,guesses,answers)
    print("Thanks for playing!")
    
def printboard(clue, guesses, answers):
    for word in answers:
        if word not in guesses:
            print(len(word)*"_ ")
        else:
            print(word)
    print(''.join(clue))
    
def getallwords(clue):
    p = []
    for i in range(3, 8):
        p+=getwords(clue, i)
    return p

def getwords(clue, n):
    words = list(permutations(clue,n))
    p = [''.join(w) for w in words]
    p = [w for w in p if w in WORDS]
    p = sorted(list(set(p)))
    return p


main()
    
```
